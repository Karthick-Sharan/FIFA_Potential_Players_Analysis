# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
import warnings

warnings.filterwarnings('ignore')

sns.set_style(style='darkgrid')
```

# Importing Data

```python
df = pd.read_csv('fifa.csv')
dff = pd.read_csv('fifa_ variable_information.csv')
dff
```

```python
df.columns
```

# Dropping Columns which are not needed

```python
df.drop(columns=['Photo','Flag','Club Logo','Jersey Number'], inplace=True)
```

```python
df
```

# Treating 'Player Value' Column

```python
df['Value'] = df['Value'].apply(lambda x: x.strip('€'))
df['Value'] = df['Value'].apply(lambda x : float(x.strip('M'))*1000000 if x.find('K')==-1 else float(x.strip('K'))*1000)
df['Value'] = df['Value'].astype('float')
```

```python
df['Value']
```

# Treating 'Weekly Wage' Column

```python
df['Wage'] = df['Wage'].apply(lambda x: x.strip('€'))
df['Wage'] = df['Wage'].apply(lambda x : float(x.strip('M'))*1000000 if x.find('K')==-1 else float(x.strip('K'))*1000)
df['Wage'] = df['Wage'].astype('float')
```

```python
df['Wage']
```

# Treating 'Joing Date' of Player

```python
df['Joined'] = df['Joined'].fillna('0 0')
#Club Joining date of a player is not really important to us so we're setting this as '0' (an arbitrary values)
df['Joined'].isnull().sum()
```

```python
df['Joined'] = df['Joined'].apply(lambda x:x.split()[-1])
df['Joined'] = df['Joined'].astype(int)
```

```python
df['Joined']
```

# Treating 'Current contract validity' of Player

```python
df['Contract Valid Until'] = df['Contract Valid Until'].fillna('2099')
#Since we don't know player's contract expiry date, we're assuming he won't be becoming a free agent anytime soon,
# so setting his contract date till year 2099 
df['Contract Valid Until'] = pd.to_datetime(df['Contract Valid Until'])
```

```python
df['Contract Valid Until']
```

# Imputing missing 'Height' based on player's 'Balance'

```python
ind = df[df['Height'].isnull()].index
b = 'Balance'
# We're using 'Balance' stat of player to determine height, since both of them have a strong negative correlation
for i in ind:
    df.loc[i,'Height'] = "6'2" if df.loc[i,b]<50 else "6'0" if df.loc[i,b]>=50 and df.loc[i,b]<75 else "5'8" if df.loc[i,b]>=75 and df.loc[i,b]<90 else "5'6"
```

```python
df['Height'] = df['Height'].apply(lambda x : float(x.split("'")[0])*0.3048 + float(x.split("'")[1])*0.0254 )
df['Height']
```

# Treating missing values for 'Weight'

```python
df[df['Weight'].isnull()]['Overall'].max()/len(df)*100, df[df['Weight'].isnull()]['Overall'].max()
```

```python
df = df.dropna(subset=['Weight'])  
#Dropping rows with NaN in Weight since it's only 0.3% of total Data, and Max Overall within that subset is only 62 
```

```python
df[df['Weight'].isnull()]
```

```python
df['Weight'] = df['Weight'].apply(lambda x : x.strip('lbs'))
df['Weight'] = df['Weight'].astype(float)
```

```python
df['Weight']
```

# Treating 'Release Clause' column

```python
df['Release Clause'] = df['Release Clause'].fillna('€999M')
# Since these players don't have a release clause we can't approach them directly and have to negotiate via their club, 
# so we're setting it to 999M, since it's not possible to directly go to contract negitiation with the player,
# it's the same as being unable to afford the release clause.
```

```python
df['Release Clause'] = df['Release Clause'].apply(lambda x: x.strip('€'))
df['Release Clause'] = df['Release Clause'].apply(lambda x : float(x.strip('M'))*1000000 if x.find('K')==-1 else float(x.strip('K'))*1000)
df['Release Clause'] = df['Release Clause'].astype('float')
```

```python
# 4. Generate tables containing the top 5 players by Overall rating for each unique position.
#    a) Are there any players appearing in more than one Table. Please point out such players.
#    b) What is the average wage one can expect to pay for the top 5 in every position?
```

```python
df['Position'].isnull().sum(), df[df['Position'].isnull()]['Overall'].max()
```

```python
# Since there are only 12 players with missing position info, and the max Overall among those players is 70,
# we can simply drop these rows as they won't affect out final result.
df.dropna(subset=['Position'], inplace=True)
```

```python
df['Position'].isnull().sum()
```

```python
pos = df['Position'].unique()
pos
```

```python
df_dict = {}
for i in pos:
    df_dict[i] = df[df['Position']==i].sort_values(by='Overall', ascending=False).head(5)
    
#creating a dictionary where key is 'Position' and value is the DF of top 5 players in that position
```

```python
# Randomly testing positions : 
df_dict['ST']
```

```python
df_dict['GK']
```

```python
# 4.a
l=[]
m=[]
common = []

for j in pos:
    m = l
    l =[]
    for i in df_dict[j].index:
        l.append(i)
    for k in set(l)&set(m):
        common.append(k)

print('Number of Players who show up in more than one table is :',len(common))
```

```python
# 4.b
for i in df_dict:
    print('Average Wage for top 5 players in position', i, "is : €", df_dict[i]['Wage'].mean())
```
